CSV files from Kantar cannot be used off the bat, these ones are particularly large so one must remove the excess text above the headers in order to load it properly into a data frame.

Best to use notepad for this. Excel will trunkcate the data when saving.

Also because the file is large, the cell to load it into a df will take some time.

```python
fpath = path + "/" + fnames[5]
fpath
```

```python
df = pd.read_csv(fpath)
```

# DO NOT RUN THE ABOVE CELL A SECOND TIME
# IT READS LOTS OF DATA

```python
df.head(10)
```

```python
df1 = df[:-3] # file excludes all excess data
df1.tail()
```

```python
df1.describe()
```

```python
for key in df1.keys():
    newkey = key.replace(" ","")
    df1[newkey] = df1.pop(key) 
    # entries are items, removes entry for that key and replaces old key with 
    # new name and attaches old vals'
    
print(df1.keys())
```

```python
dp = []
for daypart in df1['DAYPART']:
    if 'Daytime' in daypart:
        dp.append('Daytime')
    else:
        dp.append(daypart)
        
```

```python
import collections as cl # check counts
counts = cl.Counter(dp)
print(counts)
```

```python
df1['dp'] = dp
```

```python
dfx = df1.loc[:,['PROPERTY','dp', 'W50-64GRP']]
dfx # experimental
```

## The below pivot table is important, it reflects the desired outcome table
Note: Remove all dp except for Daytime and Prime 

```python
pd.pivot_table(dfx, index = ['PROPERTY', 'dp'], aggfunc = 'mean')
```

```python
dfx.pivot_table(index = ['PROPERTY', 'dp'], aggfunc = 'mean')
```

```python
dfx_1 = dfx[dfx.dp == 'Daytime' ]
dfx_1
```

```python
dfx_2 = dfx[dfx.dp == 'Prime' ]
dfx_2
```

```python
dfx_all = pd.concat([dfx_1, dfx_2])
dfx_all
```

## Below is the final table to export

```python
dfx_all.pivot_table(index = ['PROPERTY', 'dp'], aggfunc = 'mean')
```
